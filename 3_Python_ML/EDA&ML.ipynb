```python
###revenue forcasting
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
```

```python
monthly_sales = (
    df.groupby(['Year', 'Month'])
      .agg({'Revenue': 'sum'})
      .reset_index()
)

monthly_sales.head()
```

```python
X = monthly_sales[['Year', 'Month']]
y = monthly_sales['Revenue']
```

```python
from sklearn.linear_model import LinearRegression

rev_model = LinearRegression()
rev_model.fit(X, y)
```

```python
from sklearn.metrics import mean_absolute_error

pred = rev_model.predict(X)
mae = mean_absolute_error(y, pred)
mae
```

```python
###churn prediction
```

```python
latest_date = df['OrderDate'].max()
df['RecencyDays'] = (latest_date - df['OrderDate']).dt.days
```

```python
df['Churn'] = df['RecencyDays'].apply(lambda x: 1 if x > 90 else 0)
```

```python
customer_df = (
    df.groupby('CustomerKey')
      .agg({
          'Revenue': 'sum',
          'OrderQuantity': 'sum',
          'RecencyDays': 'min',
          'Churn': 'max'
      })
      .reset_index()
)

customer_df.head(25)
```

```python
from sklearn.ensemble import RandomForestClassifier

X = customer_df[['Revenue', 'OrderQuantity', 'RecencyDays']]
y = customer_df['Churn']

churn_model = RandomForestClassifier(random_state=42)
churn_model.fit(X, y)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, churn_model.predict(X)))
```
